```python
import pandas as pd
import re
from transformers import AutoTokenizer, AutoModel
import torch
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from scipy.spatial.distance import cdist
import numpy as np
import matplotlib.pyplot as plt

def preprocess_text(text):
    text = re.sub(r'[^a-zA-ZçğıöşüÇĞİÖŞÜ\s]', '', str(text))
    text = text.lower()
    return text

def generate_embeddings(texts, model, tokenizer):
    embeddings = []
    for text in texts:
        inputs = tokenizer(text, return_tensors='pt', truncation=True, padding=True, max_length=128)
        outputs = model(**inputs)
        embedding = outputs.last_hidden_state.mean(dim=1).squeeze().detach().numpy()
        embeddings.append(embedding)
    return torch.tensor(embeddings)

def validate_clusters(embeddings, labels, kmeans):
    silhouette_avg = silhouette_score(embeddings, labels)
    cluster_distances = []
    for i in range(kmeans.n_clusters):
        cluster_points = embeddings[labels == i]
        if len(cluster_points) > 1:
            intra_distances = cdist(cluster_points, [kmeans.cluster_centers_[i]])
            avg_distance = np.mean(intra_distances)
            cluster_distances.append(avg_distance)
        else:
            cluster_distances.append(0)
    inter_cluster_distances = cdist(kmeans.cluster_centers_, kmeans.cluster_centers_)
    np.fill_diagonal(inter_cluster_distances, np.nan)
    metrics = {
        'Silhouette Score': silhouette_avg,
        'Average Intra-Cluster Distance': np.mean(cluster_distances),
        'Minimum Inter-Cluster Distance': np.nanmin(inter_cluster_distances),
        'Average Inter-Cluster Distance': np.nanmean(inter_cluster_distances)
    }
    return metrics

def test_and_plot_clusters(file_path, cluster_list):
    df = pd.read_excel(file_path)
    df['Processed_Text'] = df['Text'].apply(preprocess_text)
    model_name = "dbmdz/bert-base-turkish-cased"
    tokenizer = AutoTokenizer.from_pretrained(model_name)
    model = AutoModel.from_pretrained(model_name)
    embeddings = generate_embeddings(df['Processed_Text'].tolist(), model, tokenizer).numpy()
    results = []
    best_df = None
    best_silhouette = -1
    for n_clusters in cluster_list:
        kmeans = KMeans(n_clusters=n_clusters, random_state=0)
        labels = kmeans.fit_predict(embeddings)
        df['Cluster'] = labels
        metrics = validate_clusters(embeddings, labels, kmeans)
        metrics['Number of Clusters'] = n_clusters
        results.append(metrics)
        print(f"Number of Clusters: {n_clusters} - Metrics: {metrics}")
        if metrics['Silhouette Score'] > best_silhouette:
            best_silhouette = metrics['Silhouette Score']
            best_df = df.copy()
    results_df = pd.DataFrame(results)
    # Plot metrics
    plt.figure(figsize=(12, 6))
    plt.plot(results_df['Number of Clusters'], results_df['Silhouette Score'], marker='o', label='Silhouette Score')
    plt.plot(results_df['Number of Clusters'], results_df['Average Intra-Cluster Distance'], marker='o', label='Average Intra-Cluster Distance')
    plt.plot(results_df['Number of Clusters'], results_df['Average Inter-Cluster Distance'], marker='o', label='Average Inter-Cluster Distance')
    plt.xlabel('Number of Clusters')
    plt.ylabel('Metric Value')
    plt.title('Clustering Validation Metrics vs. Number of Clusters')
    plt.legend()
    plt.grid(True)
    plt.show()
    # Return the best DataFrame
    return best_df

def save_best_cluster_to_excel(df, file_name):
    df.to_excel(file_name, index=False)

# Usage example:
file_path = 'resmigazete_tarım_filter.xlsx'
cluster_list = [10, 15, 20, 25, 30, 35, 40, 45, 50, 55, 60, 65, 70, 80, 90, 100, 105, 110, 120, 130, 150, 170, 180]
test_and_plot_clusters(file_path, cluster_list)
```

```python
import pandas as pd
import re
from transformers import AutoTokenizer, AutoModel
import torch
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.feature_extraction.text import TfidfVectorizer

def preprocess_text(text):
    text = re.sub(r'[^a-zA-ZçğıöşüÇĞİÖŞÜ\s]', '', str(text))
    text = text.lower()
    return text

def generate_embeddings(texts, model, tokenizer):
    embeddings = []
    for text in texts:
        inputs = tokenizer(text, return_tensors='pt', truncation=True, padding=True, max_length=128)
        outputs = model(**inputs)
        embedding = outputs.last_hidden_state.mean(dim=1).squeeze().detach().numpy()
        embeddings.append(embedding)
    return torch.tensor(embeddings)

def extract_cluster_keywords(df, num_keywords=5):
    """
    Extracts top keywords for each cluster using TF-IDF.
    
    Parameters:
    - df (pd.DataFrame): Dataframe containing text and cluster assignments.
    - num_keywords (int): Number of keywords to extract per cluster.
    
    Returns:
    - dict: Dictionary mapping cluster labels to lists of top keywords.
    """
    tfidf_vectorizer = TfidfVectorizer(max_features=1000, stop_words=None)  # Adjust language as needed
    tfidf_matrix = tfidf_vectorizer.fit_transform(df['Processed_Text'])
    
    keywords = {}
    for cluster in df['Cluster'].unique():
        cluster_texts = df[df['Cluster'] == cluster]['Processed_Text']
        tfidf_submatrix = tfidf_vectorizer.transform(cluster_texts)
        avg_tfidf_scores = np.mean(tfidf_submatrix.toarray(), axis=0)
        top_indices = avg_tfidf_scores.argsort()[-num_keywords:][::-1]
        top_terms = [tfidf_vectorizer.get_feature_names_out()[i] for i in top_indices]
        keywords[cluster] = top_terms
    
    return keywords

def cluster_texts_with_bert(file_path, num_clusters=5):
    df = pd.read_excel(file_path)
    if 'Text' not in df.columns:
        raise ValueError("The file must contain a 'Text' column.")
    df['Processed_Text'] = df['Text'].apply(preprocess_text)
    model_name = "dbmdz/bert-base-turkish-cased"
    tokenizer = AutoTokenizer.from_pretrained(model_name)
    model = AutoModel.from_pretrained(model_name)
    embeddings = generate_embeddings(df['Processed_Text'].tolist(), model, tokenizer)
    kmeans = KMeans(n_clusters=num_clusters, random_state=0)
    df['Cluster'] = kmeans.fit_predict(embeddings)
    pca = PCA(n_components=2)
    df['PCA1'] = pca.fit_transform(embeddings)[:, 0]
    df['PCA2'] = pca.fit_transform(embeddings)[:, 1]
    cluster_keywords = extract_cluster_keywords(df)
    df['Cluster_Keywords'] = df['Cluster'].map(cluster_keywords)
    return df, cluster_keywords

# Usage example:
file_path = 'resmigazete_tarım_filter.xlsx'
num_clusters = 50  # Define number of clusters as needed
processed_df, cluster_keywords = cluster_texts_with_bert(file_path, num_clusters)

# Sort and save the processed dataframe with cluster info and keywords
processed_df = processed_df.sort_values(by='Cluster')
processed_df.to_excel("BERT_cluster.xlsx", index=False)
```
