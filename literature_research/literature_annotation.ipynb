## Literature Annotation Model

Literature annotation plays an important part in forming the study. The researcher, in the beginning, creates a topic abstraction model based on their followed readings and the intention of the research project. The topic structure changes during this process, involving defining, merging, and eliminating the topics that are initially thought of. Hypotheses and arguments are formed by these presumed topics, followed by readings, and the researcher's framework.

Here I defined my initial recuperées for later referencing how my thought process changed. There are some parameters for topic abstraction.

1. I am trying to keep topic number limited and avoid overloading the annotations with topics. This bias would lead to the underspecification of some annotations.
2. Preliminary readings were done throughout the last year. I started with 42 topics for the first recording process. These topics were exported from my first 10 recorded readings using Zotero. Since then, 2 new categories have been added. 
3. To prevent underspecification, I employ less-used topics for a multi-theoretical approach to the food sovereignty. In general, I tend to use 2-3 categories for defining an annotation with the abstracted topics. I will use mutually referencing topics interchangeably, to not forget the connections in the agri-food system.
4. The third part can be considered for multiple topics at the same time, so underspecification will still occur. As said, annotations are kept under a limit of specification, but I wonder if there is a tendency to overspecify initial annotations and underspecify the last ones.

Additionally, I packed two categories for implying another category that was not created. Here I used "Variable" and "Interdisciplinary" for indicating qualitative variables. Interdisciplinary on its own shows the socio-theoretical approach for agroecology and food system research.

The use of "Food_Sovereignty" and "Food_Security" in the same annotation indicates the interplay and/or conflict between these theoretical frameworks.

For other less-apparent categories:

Food sovereignty theory emphasizes local autonomy, therefore it is both a latent and target variable.

```python
import pandas as pd

class Annotation:
    def __init__(self):
        self.Title = ""
        self.Citation = ""
        self.In_Text_Citation = ""
        self.Agro_econ = ""
        self.Agro_international = ""
        self.Agro_policy = ""
        self.Agro_tech = ""
        self.Agroecology = ""
        self.Autonomy = ""
        self.Big_agro = ""
        self.Buyuksehir_Law = ""
        self.Collectives = ""
        self.Cooperatives = ""
        self.Debt = ""
        self.Deruralization = ""
        self.Education = ""
        self.Food_Network = ""
        self.Food_Security = ""
        self.Food_Sovereignty = ""
        self.Gender = ""
        self.Health = ""
        self.History = ""
        self.Interdisciplinary = ""
        self.Land_Consolidation = ""
        self.Land_Policy = ""
        self.Land_Use = ""
        self.Methodology = ""
        self.Migration = ""
        self.Monoculture = ""
        self.Monoculture_Poli = ""
        self.Policy_Access = ""
        self.Risks_Global = ""
        self.Rural_Development = ""
        self.Rural_Family = ""
        self.Rural_Livelihood = ""
        self.Rural_Policy = ""
        self.Rurban = ""
        self.Seed = ""
        self.Shortfood = ""
        self.Small_holder = ""
        self.Survivorship_bias = ""
        self.TR_agroEcon = ""
        self.TR_agroGov = ""
        self.TR_landUse = ""
        self.TR_ruralGov = ""
        self.TR_Peasant = ""
        self.Urbanization = ""
        self.Variable = ""

    def add_in_text_citation(self, citation):
        self.In_Text_Citation = f"{citation}"

    def add_citation(self, citation):
        self.Citation = f"{citation}"
        
    def add_title(self, title):
        self.Title = f"{title}"

    def add_annotation(self, attribute, annotation):
        if hasattr(self, attribute):
            current_value = getattr(self, attribute)
            if current_value:
                updated_value = f'{current_value},"{annotation}"'
            else:
                updated_value = f'"{annotation}"'
            setattr(self, attribute, updated_value)
        else:
            raise AttributeError(f"Attribute {attribute} does not exist in Annotation class.")

    def add_to_excel(self, filename):
        # Load existing Excel file
        df = pd.read_excel(filename)

        # Add a new row for the current instance
        new_row = {attr: getattr(self, attr) or 0 for attr in self.__dict__.keys()}
        df = pd.concat([df, pd.DataFrame([new_row])], ignore_index=True)

        # Save back to Excel file
        df.to_excel(filename, index=False)



def add_annots_from_dict(annotation_obj, annotations_dict):
    """
    Add annotations to the provided Annotation object from a dictionary.
    
    Parameters:
    - annotation_obj: Instance of Annotation class
    - annotations_dict: Dictionary where keys are annotation texts and values are lists of categories
    """
    for annotation_text, categories in annotations_dict.items():
        for category in categories:
            annotation_obj.add_annotation(category, annotation_text)

```
